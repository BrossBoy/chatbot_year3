# scraping data

```python
def scraping():
    url = 'https://bester-eats.com/area/PRO64/ARE829/LCAT10/CAT1015/'

    # set up the webdriver
    driver = webdriver.Chrome(options=chrome_options)
    driver.get(url)

    element = driver.find_elements(By.XPATH, "//a[@class='restaurant-card-header']")
    url = [str(i.get_attribute('href')) for i in element]
    list_name = [str(i.text) for i in element]

    list_address = []
    list_phone = []
    list_open = []
    list_end = []
    list_price = []
    for i in url:
        driver.get(i)

        address = driver.find_element(By.XPATH, "//a[@class='restaurant-address__link gtm--basic-info-address']").text

        phone = driver.find_element(By.XPATH, "//span[@class='restaurant-tel__num']")

        open_time = driver.find_element(By.XPATH, "//td[@class='restaurant-info__value']/dl/dd").text
        open_time = re.findall('\d+', open_time)
        open = int(open_time[0]+open_time[1])/100
        end = int(open_time[2]+open_time[3])/100

        tmp = driver.find_elements(By.XPATH, "//span[@class='cost-item__price cost-item__price--bottom']")
        price = ' '.join([str(j.text) for j in tmp])
        price = int(re.findall('\d+', price)[0])

        list_address.append(address)
        list_phone.append('0'+phone.text[3:])
        list_open.append(open)
        list_end.append(end)
        list_price.append(price)

    return list_name, list_address, list_phone, list_open, list_end, list_price
```

```python
def detail(s, i):
    text = " "
    if s[0]:
      text += address[i]+" "
    if s[1]:
      text += phone[i]
    return text+"\n"

def final_ans(pre_ans):
    ans_text = ""
    time_recommen = []
    price_recommen = []
    time_ref_recommen = []
    price_ref_recommen = []
    double_check = -1
    if not pre_ans[0] is None:
      ans_text += "* ร้านที่คุณตามหา "+ name[pre_ans[0]]+ detail(pre_ans[3:], pre_ans[0])
      double_check = pre_ans[0]
      if pre_ans[1] is None and pre_ans[2] is None:
          for i in range(len_dimsom):
            if open[i] <= open[pre_ans[0]] and close[pre_ans[0]] <= close[i] and i != double_check:
              # print(name[i], open[i], close[i])
              time_ref_recommen.append(i)
            if price[i] <= price[pre_ans[0]] and i != double_check:
              # print(name[i], price[i])
              price_ref_recommen.append(i)
          ref_recommen = list(set(time_ref_recommen).intersection(price_ref_recommen))
          # print(ref_recommen)
          if len(ref_recommen) != 0:
            ans_text += "\n* ร้านที่ราคาและเวลาใกล้เคียงกับร้านที่คุณสนใจ\n"
            for i in range(len(ref_recommen)):
              if i == 5:
                break
              ans_text +="- "+ name[ref_recommen[i]] + detail(pre_ans[3:], ref_recommen[i])
          return ans_text

    if not pre_ans[1] is None:
      for i in range(len_dimsom):
        if open[i] <= pre_ans[1] and pre_ans[1] <= close[i] and i != double_check:
          # print(name[i], open[i], close[i])
          time_recommen.append(i)

    if not pre_ans[2] is None:
      for i in range(len_dimsom):
        if price[i] <= pre_ans[2] and i != double_check:
          # print(name[i], price[i])
          price_recommen.append(i)

    if not pre_ans[1] is None and not pre_ans[2] is None:
        ans_text += f"* แนะนำร้านที่ตรงกับเกณฑ์เวลา {pre_ans[1]} นาฬิกา และราคา {pre_ans[2]} บาท\n"
        recommen = list(set(time_recommen).intersection(price_recommen))
    elif not pre_ans[1] is None:
      ans_text += f"* แนะนำร้านที่เข้าเกณฑ์เวลา {pre_ans[1]} นาฬิกา\n"
      recommen = time_recommen
    else:
      ans_text += f"* แนะนำร้านที่เข้าเกณฑ์ราคา {pre_ans[2]} บาท\n"
      recommen = price_recommen

    if len(recommen) != 0:
      for i in range(len(recommen)):
        if i == 5:
          break
        ans_text +="- "+ name[recommen[i]] + detail(pre_ans[3:], recommen[i])
    elif pre_ans[0] is None:
        ans_text += "* ขอโทษด้วยไม่สามารถหาร้านที่เข้าเกณ์ของคุณ"
    else:
        ans_text += "* ไม่สามารถหาร้านแนะนำเพิ่ม"

    return ans_text

print(final_ans([3,None,150,True,True]))
```

# ask user

```python
ask_time = -1
all_ask = ''
pre_ans = []

def chatbot(input, history=[]):
    output = get_answer(input)
    history.append((input, output))
    return history, history

def get_answer(question):
    global all_ask
    global ask_time
    global pre_ans
    if question == '.' and ask_time != -1 and ask_time < 3:
      all_ask += question
      if all_ask == "...":
        pre_ans = []
        all_ask = ""
        ask_time = 0
        return "ข้อมูลไม่เพียงพอ โปรดตอบคำถามใหม่\n"+quest[0][7:]
      pre_ans.append(None)
    elif ask_time == 0:
      quest_vec = model.encode(question, normalize_embeddings=True)
      score = list(util.cos_sim(quest_vec,name_vec))
      if max(score) >= 0.8:
        pre_ans.append(score.index(max(score)))
      else:
        return "ขออภัย ชื่อร้านที่คุณให้มาไม่มีอยู่ในรายการของผมผมขอชื่อใหม่อีกครั้ง\nตัวอย่าง โกอ้วน\nไม่ต้องการพิมพ์ ."
    elif ask_time == 1:
      num = re.findall('\d+', question)
      if len(num) == 0:
        return "ผมหาตัวเลขไม่เจอ โปรดระบุตัวเลขเพื่อบ่งบอกช่วงเวลาใหม่\nตัวอย่าง 6 โมง\nไม่ต้องการพิมพ์ ."
      elif int(num[0]) < min(open) or int(num[0]) > max(close):
        return f"ร้านเปิดช่วง {min(open)}-{max(close)} โปรดระบุตัวเลขใหม่ ไม่ต้องการพิมพ์ ."
      pre_ans.append(int(num[0]))
    elif ask_time == 2:
      num = re.findall('\d+', question)
      if len(num) == 0:
        return "ผมหาตัวเลขไม่เจอ โปรดระบุตัวเลขเพื่อบ่งบอกราคาใหม่\nตัวอย่าง 300 บาท\nไม่ต้องการพิมพ์ ."
      elif int(num[0]) < min(price):
        return f"ราคาขั้นต่ำอยู่ที่ {min(price)} บาท โปรดระบุราคาใหม่ ไม่ต้องการพิมพ์ ."
      pre_ans.append(int(num[0]))
    elif ask_time != -1:
      if question == '.':
        pre_ans.append(False)
      else:
        quest_vec = model.encode(question, normalize_embeddings=True)
        score = list(util.cos_sim(quest_vec,yes_vec))
        if max(score) >= 0.8:
          pre_ans.append(True)
        else:
          return "โปรดระบุความต้องการให้ถูกต้อง\nตัวอย่าง ใช่\nไม่ต้องการพิมพ์ ."
    ask_time += 1
    if ask_time == len(quest):
      ans = final_ans(pre_ans)
      pre_ans = []
      all_ask = ""
      ask_time = -1
      return ans
    return(quest[ask_time])
```
