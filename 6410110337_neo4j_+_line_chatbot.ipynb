```python
cypher_query = ['MATCH (n) return n.name as name;', 'MATCH ()-[n]->() return n.name as name;']
# "MATCH (n)-[r]-(p) WHERE n.name =~ 'One Piece.*' AND r.name =~ '.*นักแสดงนำ.*' RETURN p.name as name"
Entity_corpus = []
with driver.session() as session:
    results = session.run(cypher_query[0])
    for record in results:
        Entity_corpus.append(record['name'])
    # using set() to remove duplicated from list
    Entity_corpus = list(set(Entity_corpus))
print(Entity_corpus)

Relation_corpus = []
with driver.session() as session:
    results = session.run(cypher_query[1])
    for record in results:
      if record['name'] != None:
        Relation_corpus.append(record['name'])
    # using set() to remove duplicated from list
    Relation_corpus = list(set(Relation_corpus))
print(Relation_corpus)

Core_Relation_corpus = []
with driver.session() as session:
    results = session.run("MATCH (n)-[r]->() WHERE n.name='สโมสรฟุตบอลลิเวอร์พูล' RETURN r.name as name;")
    for record in results:
      Core_Relation_corpus.append(record['name'])
    # using set() to remove duplicated from list
    Core_Relation_corpus = list(set(Core_Relation_corpus))
print(Core_Relation_corpus)

Data_Entity_corpus = []
with driver.session() as session:
    results = session.run("MATCH (n) WHERE n.class='data' RETURN n.name as name;")
    for record in results:
      Data_Entity_corpus.append(record['name'])
    # using set() to remove duplicated from list
    Data_Entity_corpus = list(set(Data_Entity_corpus))
print(Data_Entity_corpus)
```

```python
from sentence_transformers import SentenceTransformer, util

# Load the BERT model. Various models trained on Natural Language Inference (NLI) https://github.com/UKPLab/sentence-transformers/blob/master/docs/pretrained-models/nli-models.md and
# Semantic Textual Similarity are available https://github.com/UKPLab/sentence-transformers/blob/master/docs/pretrained-models/sts-models.md

model = SentenceTransformer('sentence-transformers/paraphrase-multilingual-mpnet-base-v2')
entity_vec = model.encode(Entity_corpus, convert_to_tensor=True,normalize_embeddings=True)
relation_vec = model.encode(Relation_corpus, convert_to_tensor=True,normalize_embeddings=True)
core_relation_vec = model.encode(Core_Relation_corpus, convert_to_tensor=True,normalize_embeddings=True)
```

#quick reply

```python
def quick_reply_str(entity):
  # print(entity)
  str_m = []
  for i in entity:
      text = i
      str_m.append({ "type": "action","imageUrl": "","action": { "type": "message", "label": text[:18]+".." , "text": text }})
  return  {"items": str_m}

# print(quick_reply_str([3, 7]))

def relation_quick_reply_str(relation):
  # print(relation)
  str_m = []
  with driver.session() as session:
      for i in relation:
        result = session.run(f"MATCH ()-[r]->(n) WHERE r.name='{i}' RETURN n.name limit 1;")
        for record in result:
          str_m.append({ "type": "action","imageUrl": "","action": { "type": "message", "label": record['n.name'][:18]+".." , "text": record['n.name'] }})
      return  {"items": str_m}
```

# หาคำตอบ

```python
def find_ans(question):
    ask_vec = model.encode(question, convert_to_tensor=True,normalize_embeddings=True)
    cosine_score = list(util.cos_sim(entity_vec, ask_vec))
    cosine_score = [[cosine_score[i], i] for i in range(len(cosine_score))]
    cosine_score = sorted(cosine_score, reverse=True)
    max_cosine = cosine_score[0][0]

    cosine_rela_score = list(util.cos_sim(relation_vec, ask_vec))
    cosine_rela_score = [[cosine_rela_score[i], i] for i in range(len(cosine_rela_score))]
    cosine_rela_score = sorted(cosine_rela_score, reverse=True)
    max_rela_cosine = cosine_rela_score[0][0]

    # print(max_cosine, max_rela_cosine, Entity_corpus[cosine_score[0][1]], Relation_corpus[cosine_rela_score[0][1]])
    ans = ""
    if max_cosine > 0.77:
        if Entity_corpus[cosine_score[0][1]] == 'สโมสรฟุตบอลลิเวอร์พูล':
            cosine_core_rela_score = list(util.cos_sim(core_relation_vec, ask_vec))
            quest = f"MATCH (n)-[r]->(p) WHERE n.name = 'สโมสรฟุตบอลลิเวอร์พูล' and r.name = '{Core_Relation_corpus[cosine_core_rela_score.index(max(cosine_core_rela_score))]}' RETURN n.name, r.name, p.name limit 5"
            with driver.session() as session:
              results = session.run(quest)
              for i in results:
                ans += f"{i['n.name']}{i['r.name']}{i['p.name']}\n"

            pre_quick = []
            for i in range(1, len(cosine_score)):
                if Entity_corpus[cosine_score[i][1]] in Data_Entity_corpus:
                  pre_quick.append(Entity_corpus[cosine_score[i][1]])
                if len(pre_quick) == 3:
                  break

            return ans[:-1]+"\nสิ่งที่คุณอาจจะสนใจ", quick_reply_str(pre_quick)
        else:
            quest = f"MATCH (n)-[r]->(p) WHERE n.name = '{Entity_corpus[cosine_score[0][1]]}' or p.name = '{Entity_corpus[cosine_score[0][1]]}' RETURN n.name, n.class, r.name, p.name limit 5"
            with driver.session() as session:
              results = session.run(quest)
              results = list(results)
              for i in results:
                if i["n.class"] == "greeting":
                  return i["p.name"], ""
                elif i['n.name'] != 'สโมสรฟุตบอลลิเวอร์พูล' or len(results) == 1:
                  ans += f"{i['n.name']}{i['r.name']}{i['p.name']}\n"

              pre_quick = []
              for i in range(1, len(cosine_score)):
                  if Entity_corpus[cosine_score[i][1]] in Data_Entity_corpus:
                    pre_quick.append(Entity_corpus[cosine_score[i][1]])
                  if len(pre_quick) == 3:
                    break

              return ans[:-1]+"\nสิ่งที่คุณอาจจะสนใจ", quick_reply_str(pre_quick)
    elif max_rela_cosine > 0.6:
        quest = f"MATCH (n)-[r]->(p) WHERE r.name = '{Relation_corpus[cosine_rela_score[0][1]]}' RETURN n.name, r.name, p.name limit 5"
        with driver.session() as session:
            results = session.run(quest)
            for i in results:
              ans += f"{i['n.name']}{i['r.name']}{i['p.name']}\n"

            pre_quick = [Relation_corpus[cosine_rela_score[1][1]], Relation_corpus[cosine_rela_score[2][1]], Relation_corpus[cosine_rela_score[3][1]]]

            return ans[:-1]+"\nสิ่งที่คุณอาจจะสนใจ", relation_quick_reply_str(pre_quick)
    else:
      return "ขอโทษด้วยผมไม่สามารถเข้าใจสิ่งที่คุณกล่าวมาก โปรดถามผมเกี่ยวกับข้อมูลเบื้องต้นของสโมสรลิเวอร์พูล", ""
```

```python
app = Flask(__name__)

@app.route("/", methods=['POST'])
def linebot():
    body = request.get_data(as_text=True)                    # รับเนื้อหาของข้อความที่ได้รับ
    try:
        json_data = json.loads(body)                         # json จัดรูปแบบเนื้อหาข้อความ
        access_token = 'hrTe7ZKiWpf4K/XLLMBlPEwVStpBiwmEXkrZZh9KZRt5PxcDLDqvi9fuyXfjg6viEiNtCJWJ+HhxD0oVbV04jE5Ni6ATxOLEGPY2pLJDjDEEs10s75ytf0FQ5gVGz4gVP3H2GexjWMx3eeirfWSfYQdB04t89/1O/w1cDnyilFU='
        secret = '97f2a1056d01fd19faedd202b1ae5414'
        line_bot_api = LineBotApi(access_token)              # ยืนยันว่าโทเค็นถูกต้องหรือไม่
        handler = WebhookHandler(secret)                     # ยืนยันว่าความลับถูกต้องหรือไม่
        signature = request.headers['X-Line-Signature']      # เพิ่มไปยัง postback แล้ว headers
        handler.handle(body, signature)                      # ข้อมูลที่เกี่ยวข้องกับการส่งคืนข้อความการผูก
        ask = json_data['events'][0]['message']['text']      # รับข้อความที่ได้รับจาก LINE
        ans = find_ans(ask)
        tk = json_data['events'][0]['replyToken']            # รับข้อความตอบกลับ Token
        line_bot_api.reply_message(tk,TextSendMessage(text=ans[0],quick_reply=ans[1]))  # ข้อความส่งคืน
        print(ask, tk)                                       # พิมพ์เนื้อหา
    except:
        print(body)                                          # หากเกิดข้อผิดพลาด ให้พิมพ์สิ่งที่ได้รับออกมา
    return 'OK'                 # ตรวจสอบการใช้งาน Webhook ไม่สามารถละเว้นได้
if __name__ == "__main__":
  run_with_ngrok(app)           # เชื่อมต่อบริการ ngrok
  app.run()
```
